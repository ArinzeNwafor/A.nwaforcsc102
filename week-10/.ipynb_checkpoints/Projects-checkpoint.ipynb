```python
# Project 1 
# Program that allows customers to buy oranges online

class OrangeShop:
    def __init__(self, stock):
        self.stock = stock
    
    def buy_oranges(self, quantity):
        self.quantity = quantity
        if quantity <= self.stock:
            cost = quantity * 50  # Assuming each orange costs $0.5
            print(f"You have bought {quantity} oranges.")
            print(f"The total cost is ₦{cost:.2f}.")
            print("Thank you for your patronage!")
        else:
            print("Sorry, we don't have enough oranges in stock.")
            print(f"Please select a quantity less than {quantity}.")


# Main program
stock_quantity = 50  # Initial stock quantity
orange_shop = OrangeShop(stock_quantity)
print("Welcome to the Orange Shop!")
print(f"We currently have {stock_quantity} oranges in stock.")
quantity = int(input("How many oranges would you like to buy? "))
orange_shop.buy_oranges(quantity)
```
